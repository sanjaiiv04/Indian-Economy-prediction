```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/indianEco.csv')
```

```python
df.head()
```

```python
df['Target']=df['GDP growth (annual %)']
```

```python
df=df.drop(columns=['GDP growth (annual %)'],axis=1)
```

```python
df=df.drop(columns=['Country Name'],axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
x=df.drop('Target',axis=1)
y=df['Target']
```

```python
for i in x.columns:
  x[i]=scaler.fit_transform(x[[i]])
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
clf=LinearRegression().fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
```

To calculate the accuracy we can use Mean square error, Mean absolute error and Root mean square error

```python
from sklearn.metrics import mean_squared_error
```

```python
print("MSE=",mean_squared_error(y_test,y_pred))
```

```python
print("RSM=",np.sqrt(mean_squared_error(y_test,y_pred)))
```
